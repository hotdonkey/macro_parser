```python
import pandas as pd
import requests
import numpy as np


def kitco_db():
    url = 'https://www.kitco.com/gold.londonfix.html'
    response = requests.get(url, headers={'User-Agent': 'Mozilla/5.0'})

    kitco_response = pd.read_html(response.text)

    kitco_df = kitco_response[1]
    kitco_df.drop([1, 4, 6], axis=1, inplace=True)
    kitco_df.columns = kitco_df.iloc[0]
    kitco_df.drop([0, 1, 2], axis=0, inplace=True)

    kitco_df['Date'] = pd.to_datetime(kitco_df['Date'])
    kitco_df = kitco_df.replace({'-': np.nan})
    kitco_df = kitco_df.sort_values(by=['Date'])
    kitco_df = kitco_df.reset_index(drop=True)
    kitco_df[['Gold', 'Silver', 'Platinum', 'Palladium']] = kitco_df[[
        'Gold', 'Silver', 'Platinum', 'Palladium']].apply(pd.to_numeric)

    kitco_df.drop_duplicates(inplace=True)

    kitco_df.to_excel(
        '/Users/hotdonkey/Documents/GitHub/data_science_learning/code/LME/parser_beta/data/kitko_db.xlsx', sheet_name='kitco_metall')
    return kitco_df
```

```python
kitco = kitco_db()
```
